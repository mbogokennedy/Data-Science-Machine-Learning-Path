```python
import pandas as pd
import numpy as np
from plotly.figure_factory import create_distplot
import plotly.graph_objs as go
import statsmodels.api as sm

from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot
init_notebook_mode()

df = pd.read_csv('modified_25.csv')
df.head()
df.shape
ds= df.replace(np.nan, 0)
ds.head()

def coun(a, b, c):
    for i in a:
        dc = b[b['Country'] == i]
        X = dc[['Emp_modelled', 'GFCFpc_Log', 'Rd']]
        X = sm.add_constant(X)
        y = dc[['GDPpc_Log']]
        model = sm.OLS(y, X).fit()
        rd_coeff = model.params[2]
        c.append(rd_coeff)
    
# Sub - saharan africa
dSS = ds[ds['Region'] == 'Sub-Saharan Africa']

ssh_countries = dSS.Country.unique().tolist()

ls = []

coun(ssh_countries, dSS, ls)


# Latin America & Caribbean
dLA = ds[ds['Region'] == 'Latin America & Caribbean']

la_countries = dLA.Country.unique().tolist()

lsla = []
coun(la_countries, dLA, lsla)

Sub_Saharan_Africa = go.Histogram(
    x=ls,
    opacity=0.75,
    name = '<b>Sub-Saharan Africa</b>',
    histfunc= 'count',
    hoverinfo= 'all',
    showlegend = True
)
Latin_America_Caribbean = go.Histogram(
    x=lsla,
    opacity=0.75,
    name = '<b>Latin America & Caribbean</b>',
    histfunc= 'count',
    hoverinfo = 'all',
    showlegend = True
)

data = [Sub_Saharan_Africa, Latin_America_Caribbean]
layout = go.Layout(barmode='overlay', 
                title = go.layout.Title(text='<b>Sub-Saharan Africa vs Latin America & Caribbean</b>' ),
                xaxis = go.layout.XAxis(title='<b>Size of the Coeficient</b>'),
                yaxis= go.layout.YAxis(title='<b>Number of the Countries</b>'))
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename='Continent Histogram')
```

```python
# Middle East & North Africa
dNM = ds[ds['Region'] == 'Middle East & North Africa']

mENA_countries = dNM.Country.unique().tolist()

lsme = []
coun(mENA_countries, dNM, lsme)
Middle_East_North_Africa = go.Histogram(
    x=lsme,
    opacity=0.75,
    name = '<b>Middle East & North Africa</b>',
    histfunc= 'count',
    hoverinfo = 'all',
    showlegend = True
)

data = [Sub_Saharan_Africa, Middle_East_North_Africa]
layout = go.Layout(barmode='overlay', 
                title = go.layout.Title(text='<b>Sub-Saharan Africa vs Middle East & North Africa</b>' ),
                xaxis = go.layout.XAxis(title='<b>Size of the Coeficient</b>'),
                yaxis= go.layout.YAxis(title='<b>Number of the Countries</b>'))
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename='Continent Histogram')
```
